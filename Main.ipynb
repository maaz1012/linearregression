```python
x = Data[['TV', 'Radio', 'Newspaper']]
y = Data[["Sales"]]
```

```python
# Training
# Train And test split

# 70% - 80% Training
# 20% - 30% Testing
```

```python
# y = b0 + b1x1 + b2x2 + b3x3
! pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape,"Training data size for for x")
print(x_test.shape,"Teststing data size for x")
print(y_train.shape,"Training data sixe for y")
print(y_test.shape,"Testing data size for y")
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

model = lm.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y = b0 + b1x1 + b2x2 + b3x3

b0 = 2.93889945
b1 = 0.04583629
b2 = 0.18611951
b3 = -0.00056731

y = b0 + b1*44.5 + b2*39.3 + b3*45.1
```

```python
y
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
new_df = pd.DataFrame()
new_df = x_test
new_df['actual sales']= y_test
new_df['predicted sales']= y_pred
new_df
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2=r2_score(y_test,y_pred)
print("R-squared:",r2)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("RMSE:",rmse)

```
